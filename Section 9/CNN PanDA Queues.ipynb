```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras.layers import Dense, Input, GlobalMaxPooling1D
from tensorflow.keras.layers import Conv1D, MaxPooling1D
from tensorflow.keras.models import Model
np.random.seed(123)
tf.random.set_seed(123)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
df = pd.read_csv('aglt2.csv', index_col='datetime', parse_dates=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df.index = pd.to_datetime(df.index, format='YYYY-MM-DD HH:MI:SS')
```

```python
df = df[['queue_fullness']].iloc[3500:3800,:].sort_index()
```

```python
df['queue_fullness'] = df['queue_fullness'].apply(lambda x: x if x > 0.0000 else 0.0001)
```

```python
df = df[~df.index.duplicated()]
```

```python
df = df.asfreq('1H')
```

```python
df = df.fillna(0.0001)
```

```python
df['LogFullness'] = np.log(df['queue_fullness'])
```

```python
df
```

```python
df.plot();
```

```python
Ntest = 6
train = df.iloc[:-Ntest]
test = df.iloc[-Ntest:]
```

```python
# boolean series to index df rows
train_idx = df.index <= train.index[-1] # indexes less than the last values of train
test_idx = df.index > train.index[-1] # indexes greater than the last value of train
```

```python
df['DiffLogFullness'] = df['LogFullness'].diff()
```

```python
df['DiffLogFullness'].plot();
```

```python
df.shape
```

```python
# make supervised dataset

series = df['DiffLogFullness'].dropna().to_numpy()

T = 10 # number of lags
X = []
Y = []
for t in range(len(series) - T):
    x = series[t:t+T]
    X.append(x)
    y = series[t+T]
    Y.append(y)

X = np.array(X).reshape(-1, T)
Y = np.array(Y)
N = len(X)
print("X.shape", X.shape, "Y.shape", Y.shape)
```

```python
Xtrain, Ytrain = X[:-Ntest], Y[:-Ntest]
Xtest, Ytest = X[-Ntest:], Y[-Ntest:]
```

```python
# CNN
i = Input(shape=(T,1))
x = Conv1D(16, 3, activation='relu')(i)
x = MaxPooling1D(2)(x)
x = Conv1D(32, 3, activation='relu')(x)
x = GlobalMaxPooling1D()(x)
x = Dense(1)(x)
model = Model(i,x)
```

```python
tf.keras.utils.plot_model(model)
```

```python
model.compile(
    loss='mse',
    optimizer='adam',
    metrics='mae'
)
```

```python
r = model.fit(
    Xtrain,
    Ytrain,
    epochs=100,
    validation_data=(Xtest, Ytest)
)
```

```python
plt.plot(r.history['loss'], label='train loss')
plt.plot(r.history['val_loss'], label='test loss')
plt.legend();
```

```python
train_idx[:T+1] = False # first T+1 values are not predictable
```

```python
Ptrain = model.predict(Xtrain).flatten() # turn to 1D array
Ptest = model.predict(Xtest).flatten()
```

```python
# Store diff predictions
df.loc[train_idx, 'Diff ANN Train Prediction'] = Ptrain
df.loc[test_idx, 'Diff ANN Test Prediction'] = Ptest
```

```python
cols = ['DiffLogFullness',
        'Diff ANN Train Prediction',
        'Diff ANN Test Prediction']
df[cols].tail(50).plot(figsize=(15,5));
```

```python
# Needed to compute un-differenced predictions
df['DiffLogFullness'] = df['LogFullness'].shift(1)
prev = df['DiffLogFullness']
```

```python
# Last-known train value
last_train = train.iloc[-1]['LogFullness']
```

```python
# 1-step forecast
df.loc[train_idx, '1step_train'] = prev[train_idx] + Ptrain
df.loc[test_idx, '1step_test'] = prev[test_idx] + Ptest
```

```python
# plot 1-step forecast
df[['LogFullness', '1step_train', '1step_test']].tail(50).plot(figsize=(15,5));
```

```python
# multi-step forecast
multistep_predictions = []

# first test input
last_x = Xtest[0]

while len(multistep_predictions) < Ntest:
    p = model.predict(last_x.reshape(1,-1))[0]

    # update the predictions list
    multistep_predictions.append(p)

    # make the new input
    last_x = np.roll(last_x, -1)
    last_x[-1] = p
```

```python
# save multistep forecast to dataframe
df.loc[test_idx, 'multistep'] = last_train + np.cumsum(multistep_predictions)
```

```python
cols = ['LogFullness', 'multistep', '1step_test']
df[cols].tail(50).plot(figsize=(15,5));
```

```python
# make multi-output supervised dataset
Tx = T # number of step in the input
Ty = Ntest # number of steps in the output
X = []
Y = []
for t in range(len(series) - Tx - Ty + 1):
    x = series[t:t+Tx]
    X.append(x)
    y = series[t+Tx:t+Tx+Ty]
    Y.append(y)

X = np.array(X).reshape(-1, Tx)
Y = np.array(Y).reshape(-1, Ty)
N = len(X)
print("X.shape", X.shape, "Y.shape", Y.shape)
```

```python
Xtrain_m, Ytrain_m = X[:-1], Y[:-1]
Xtest_m, Ytest_m = X[-1:], Y[-1:]
```

```python
# CNN
i = Input(shape=(Tx,1))
x = Conv1D(16, 3, activation='relu')(i)
x = MaxPooling1D(2)(x)
x = Conv1D(32, 3, activation='relu')(x)
x = GlobalMaxPooling1D()(x)
x = Dense(Ty)(x)
model = Model(i, x)
```

```python
model.compile(
    loss='mse',
    optimizer='adam'
)
```

```python
r = model.fit(
    Xtrain_m,
    Ytrain_m,
    epochs=100,
    validation_data=(Xtest_m, Ytest_m)
)
```

```python
plt.plot(r.history['loss'], label='train loss')
plt.plot(r.history['val_loss'], label='test loss')
plt.legend();
```

```python
Ptrain = model.predict(Xtrain_m)
Ptest = model.predict(Xtest_m)
```

```python
Ptrain.shape, Ptest.shape
```

```python
Ptrain = Ptrain[:,0]
Ptest = Ptest[0]
```

```python
# save multi-output forecast to dataframe
df.loc[test_idx, 'multioutput'] = last_train + np.cumsum(Ptest)
```

```python
# plot all forecasts
cols = ['LogFullness','multistep','1step_test','multioutput']
df[cols].tail(100).plot(figsize=(15,5));
```

```python
# MAPE
test_log_loss = df.iloc[-Ntest:]['LogFullness']
mape1 = mean_absolute_percentage_error(test_log_loss, df.loc[test_idx, 'multistep'])
print("multi-step MAPE:", mape1)

mape2 = mean_absolute_percentage_error(test_log_loss, df.loc[test_idx, 'multioutput'])
print("multi-output MAPE:", mape2)
```
