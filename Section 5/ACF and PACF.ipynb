```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xO = np.random.randn(1000)
```

```python
plt.plot(xO);
```

```python
plt.hist(xO);
```

```python
fig, ax = plt.subplots(figsize=(10,5))
plot_pacf(xO, ax=ax)
```

```python
x1 = [0]
for i in range(1000):
    x = 0.5 * x1[-1] + 0.1 * np.random.randn()
    x1.append(x)
x1 = np.array(x1)
```

```python
plt.plot(x1);
```

```python
plt.hist(x1);
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
plot_pacf(x1, ax=ax)
```

```python
errors = 0.1 * np.random.randn(1000)
ma1 = []
for i in range(1000):
    if i >= 1:
        x = 0.5 * errors[i-1] + errors[i]
    else:
        x = errors[i]
    ma1.append(x)
ma1 = np.array(ma1)
```

```python
plt.plot(ma1);
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
plot_acf(ma1, ax=ax)
```
