```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = np.random.standard_normal(1000)
```

```python
plt.plot(s)
plt.show()
```

```python
plt.hist(s)
plt.show()
```

```python
# Adding trend to the noise
line = np.linspace(0, 14, num=1000)
```

```python
s_trend = np.add(s, line)
```

```python
y = np.arange(np.min(s_trend),np.max(s_trend),(np.max(s_trend)-np.min(s_trend))/1000)
y
```

```python
plt.scatter(x=s_trend, y=y)

z = np.polyfit(s_trend, y, 1)
p = np.poly1d(z)
plt.plot(s_trend,p(s_trend),"r--")
plt.show()
```

```python
cumsum = np.cumsum(s)
```

```python
plt.plot(cumsum)
plt.show()
```

```python
from scipy.stats import multivariate_normal
x = np.random.uniform(size=(1000, 1))
y = multivariate_normal.pdf(x)
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```

```python
mean_y = 0
for i in y:
    mean_y+=i
mean_y = mean_y/1000
```

```python
mean_y
```

```python
cov_y = 0
for i in y:
    cov_y+=(i-mean_y)
cov_y = cov_y/1000
```

```python
cov_y
```

```python
np.mean(y)
```

```python
np.cov(y)
```
